**KPI Requirement**

```python
#Total Sales
Total_Sales = df['Sales'].sum()

#Average Sales
Average_Sales = df['Sales'].mean()

#No.of item sold
No_of_item_sold = df['Sales'].count()

#Average Rating
Average_Rating = df['Rating'].mean()

#Display
print(f"Total Sales: ${Total_Sales:,.0f}")
print(f"Average Sales: ${Average_Sales:,.1f}")
print(f"No. of item sold: {No_of_item_sold:,.0f}")
print(f"Average Rating: ${Average_Rating:,.1f}")
```

**Charts Requirements**

####Total Sales by Fat Content

```python
Sales_by_Fat = df.groupby('Item Fat Content')['Sales'].sum()

colors = ['#003f5c', '#ffa600']

plt.pie(Sales_by_Fat,labels=Sales_by_Fat.index,autopct='%1.1f%%', colors=colors)
plt.title('Total Sales by Fat Content', fontweight='bold')
plt.axis('equal')
plt.show()
```

####**Total sales by item type**




```python
Sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(Sales_by_item_type.index, Sales_by_item_type.values, color='#003f5c')

plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type', fontweight='bold')
plt.xticks(rotation=90)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize= 8)

plt.tight_layout()
plt.show()
```

####**Total Sales by Fat Content by Outlet**

```python
Sales_by_Fat_Outlet = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

colors = ['#003f5c', '#ffa600']
Sales_by_Fat_Outlet = Sales_by_Fat_Outlet [['Regular','Low Fat']]

Sales_by_Fat_Outlet.plot(kind='bar',  figsize=(12, 8), color=colors)
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Fat Content by Outlet', fontweight='bold')
plt.xticks(rotation=0)
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

####**Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-', color='#003f5c')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment Year', fontweight='bold')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'${y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

####**Sales by Outlet Size**

```python
sales_by_outlet_size = df.groupby('Outlet Size')['Sales'].sum()
colors = ['#003f5c', '#ffa600', '#ff7f50']
plt.figure(figsize=(4, 4))
plt.pie(sales_by_outlet_size, labels=sales_by_outlet_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Sales by Outlet Size', fontweight='bold')
plt.axis('equal')
plt.show()
```

####**Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values(by='Sales', ascending=False)

plt.figure(figsize=(8, 3))
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location , color='#003f5c')

plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.title('Total Sales by Outlet Location', fontweight='bold')

plt.tight_layout()
plt.show()

```
